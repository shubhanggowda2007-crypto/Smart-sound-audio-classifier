```python
# app.py

import streamlit as st
import sounddevice as sd
import pandas as pd
from yamnet_model import YAMNetClassifier
import time
from datetime import datetime

st.set_page_config(page_title="Audio Classification", layout="wide")
st.title("🎧 Audio Classification with YAMNet")

classifier = YAMNetClassifier()
model, class_names = classifier.load_model()

if model is None:
    st.error("Failed to load model.")
    st.stop()

option = st.radio("Choose Input Source", ("Microphone", "Audio File"))

if option == "Microphone":
    try:
        devices = sd.query_devices()
        input_devices = [dev for dev in devices if dev['max_input_channels'] > 0]
        device_names = [f"{dev['name']} (Index: {i})" for i, dev in enumerate(input_devices)]
        selected_device_name = st.selectbox("Select input microphone:", device_names)
        selected_device_index = int(selected_device_name.split('(Index: ')[1][:-1])

        if st.button("🎙 Start Recording (3 seconds)"):
            st.info("Recording...")
            duration = 3
            fs = classifier.SAMPLE_RATE
            recording = sd.rec(int(duration * fs), samplerate=fs, channels=1, dtype='float32', device=selected_device_index)
            sd.wait()
            audio_data = recording.squeeze()
            predictions = classifier.classify_audio(model, class_names, audio_data)
            fig = classifier.create_spectrogram(audio_data)

            st.subheader("Predictions:")
            df = pd.DataFrame(predictions, columns=["Class", "Confidence"])
            df["Confidence"] = df["Confidence"].apply(lambda x: f"{x * 100:.2f}%")
            st.dataframe(df)
            st.pyplot(fig)

    except Exception as e:
        st.error(f"Microphone error: {e}")

elif option == "Audio File":
    file = st.file_uploader("Upload an audio file (.wav or .mp3)", type=["wav", "mp3"])
    if file is not None:
        st.audio(file, format="audio/wav")
        if st.button("🔍 Classify Audio File"):
            predictions, audio_data = classifier.classify_audio_file(model, class_names, file)
            if predictions:
                df = pd.DataFrame(predictions, columns=["Class", "Confidence"])
                df["Confidence"] = df["Confidence"].apply(lambda x: f"{x * 100:.2f}%")
                st.dataframe(df)
                fig = classifier.create_spectrogram(audio_data)
                st.pyplot(fig)

                csv = df.to_csv(index=False).encode('utf-8')
                st.download_button("📥 Download Results", data=csv, file_name="results.csv", mime="text/csv")
            else:
                st.warning("No predictions found.")
```
